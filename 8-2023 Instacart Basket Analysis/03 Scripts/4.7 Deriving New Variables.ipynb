# 02. Deriving new variables

### PRACTICE: Sort your products into price ranges using a user-defined function

```python
# Define your function

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Check values on new column

df['price_range'].value_counts(dropna=False)
```

```python
# Check max price value

df['prices'].max()
```

### This confirms your findings from the labels—that there aren’t any high-range products within the subset.

### Create new column that summarizes how busy each day of the week is

```python
# 1. check frequency to find which day has the most orders

df_merged_final['orders_day_of_week'].value_counts(dropna=False)
```

```python
# 2. For-loop command

result =[]

for value in df_merged_final['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least day')
    else:
        result.append('Regularly day')
```

```python
# Check output
result
```

```python
# Define new column

df_merged_final['busiest_day'] = result
```

```python
# Check frequency of new column

df_merged_final['busiest_day'].value_counts(dropna=False)
```

### This shows how many sales occur on each type of day

# 03.TASK ANSWERS

```python
# create new column that summarizes the 2 busiest days & the least 2: step 1. apply For-Loops command

result_2 = []

for value in df_merged_final['orders_day_of_week']:
    if value == 0 or value == 1:
        result_2.append("Busiest days")
    elif value == 4 or value == 3:
        result_2.append("Slowest days")
    else:
        result_2.append("Regularly busy")
```

```python
# Check output
result_2
```

```python
# Create new column

df_merged_final['busiest_days'] = result_2
```

```python
# Check frequency of new column

df_merged_final['busiest_days'].value_counts(dropna=False)
```

```python
# Check output
df_merged_final['busiest_days'].info
```

### The total of 'Busiest days' corresponds to the sum of 0+1 orders (same with Slowest days: 3+4 orders) which means that Saturday&Sunday are the days with most traffic, whereas on Tuesday&Wednesday business is the slowest.

```python
# 1. Check frequency to find which time of the day registers the most orders

df_merged_final['order_hour_of_day'].value_counts(dropna=False)
```

```python
# 2.Identify the busiest hour of the day (24h grouped into 3 periods labaled as high/average/low)

result_3 = []

for value in df_merged_final['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9]:
        result_3.append("Most orders")
    elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
        result_3.append("Fewest orders")
    else:
        result_3.append("Average orders")
```

```python
result_3
```

```python
# Define the new column

df_merged_final['busiest_hours'] = result_3
```

```python
# Chech its frequency 

df_merged_final['busiest_hours'].value_counts(dropna=False)
```

```python
# Check output
df_merged_final['busiest_hours'].head
```

### This means that most orders are generated between 9AM and 4PM, whereas the traffic is lowest between 12AM and 6AM.
