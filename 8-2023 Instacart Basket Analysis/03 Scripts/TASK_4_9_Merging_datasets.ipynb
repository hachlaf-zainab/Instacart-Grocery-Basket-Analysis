## 02. Merge the two date sets

```python
# Attemp a merge

df3 = df2.merge(df1, on = ['user_id'], indicator = True)
```

```python
# Confirm merge output through merge flag frequency

df3['_merge'].value_counts(dropna = False) # result only in both, we know it´s a full-match independently.
```

```python
# Check output for new combined data set (shape, columns, head)

print(df3.shape)
```

```python
print(df3.columns)
```

```python
df3.head()
```

```python
# Drop unneccesary columns

df_final = df3.drop(columns = ['_merge','Unnamed: 0'])
```

```python
# Check output

print(df_final.shape)  # two columns dropped as intended
```

## 03. Export data

```python
# Export new dataframe as pickle file and save into my GDRIVE

df_final.to_pickle(os.path.join(path, 'final_merge.pkl'))
```
